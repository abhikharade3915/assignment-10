# converting data into horizontal data using pivot table

```python
data_book=data1.pivot_table(index='userid',columns='book',values='Rating').reset_index(drop=True)
```

```python
data_book
```

```python
data_book.index = data1.userid.unique()
```

```python
data_book
```

# Filling 'NaN' values with '0'

```python
data_book.fillna(0,inplace=True)
```

```python
data_book
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
reader_sim = 1 - pairwise_distances( data_book.values,metric='cosine')
```

```python
reader_sim
```

```python
reader_sim_df=pd.DataFrame(reader_sim)
```

```python
reader_sim_df
```
